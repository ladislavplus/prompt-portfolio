```python
import openai
import random

def run_prompt(task, user_input, target_lang=None, use_mock=False):
    """
    Send a prompt to OpenAI or simulate offline.
    """
    if use_mock:
        # offline fallback
        mock_outputs = {
            "summarize": "AI lets machines act like humans.",
            "translate": "Â¡Hola mundo!",
            "explain": "Cloud computing means using remote servers over the Internet."
        }
        return mock_outputs.get(task, "No mock result.")

    client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
    prompt = f"Task: {task}\nInput: {user_input}"
    if target_lang:
        prompt += f"\nTarget language: {target_lang}"

    response = client.chat.completions.create(
        model=os.getenv("MODEL_NAME", "gpt-4o-mini"),
        messages=[{"role": "user", "content": prompt}],
    )
    return response.choices[0].message.content.strip()

# quick test
print(run_prompt("summarize", "AI enables machines to think.", use_mock=True))
```
